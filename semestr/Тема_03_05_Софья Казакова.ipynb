## Задание 1.

Прочитать данные
`https://raw.githubusercontent.com/yakushinav/omo/main/data/movies.csv`

Имена столбцов
```
['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews', 'country',
                'cotent_rating', 'budget', 'title_year', 'imdb_score', 'genre']
```

````python
import requests
import pandas as pd
from io import StringIO

# Имена столбцов
columns = ['color', 'director_name', 'num_critic_for_reviews', 'duration',
           'gross', 'movie_title', 'num_user_for_reviews', 'country',
           'content_rating', 'budget', 'title_year', 'imdb_score', 'genre']

# Загрузка данных
url = 'https://raw.githubusercontent.com/yakushinav/omo/main/data/movies.csv'
response = requests.get(url)
data = response.text

# Создаем список словарей
movies = []
for line in data.strip().split('\n'):
    values = line.split('|')
    movie_dict = {}
    for i, col in enumerate(columns):
        if i < len(values):
            movie_dict[col] = values[i].strip()
        else:
            movie_dict[col] = None
    movies.append(movie_dict)

print(f"Загружено фильмов: {len(movies)}")
print(f"Пример: {movies[0]['movie_title']} - {movies[0]['director_name']} ({movies[0]['title_year']})")
````

## Задание 2

Прочитать данные `https://raw.githubusercontent.com/yakushinav/omo/main/data/playstore.xlsx`.

Для столбца `Last_Updated` установить формат `дата_и_время`.

````python
import pandas as pd

# Загрузка данных из Excel
url = 'https://raw.githubusercontent.com/yakushinav/omo/main/data/playstore.xlsx'
df = pd.read_excel(url)

# Преобразование столбца Last_Updated в формат дата_и_время
df['Last_Updated'] = pd.to_datetime(df['Last_Updated'])

print("Информация о данных:")
print(df.info())
print("\nПервые 5 строк:")
print(df.head())
````

## Задание 3

Прочитать данные `https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json`.

Выполнить сохранение в формат CSV. Прочитать полученный файл. Сравнить с исходным.


````python
import pandas as pd
import json
import requests

# Загрузка данных из JSON
url = 'https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json'
response = requests.get(url)
data = response.json()

# Создание DataFrame
df_json = pd.DataFrame(data)

# Сохранение в CSV
df_json.to_csv('artists.csv', index=False)

# Чтение полученного CSV файла
df_csv = pd.read_csv('artists.csv')

# Сравнение
print("=== Сравнение исходного JSON и сохраненного CSV ===")
print(f"Размер JSON данных: {df_json.shape}")
print(f"Размер CSV данных: {df_csv.shape}")
print(f"Столбцы совпадают: {list(df_json.columns) == list(df_csv.columns)}")
print(f"Данные идентичны: {df_json.equals(df_csv)}")

print("\nПервые 5 строк исходных данных:")
print(df_json.head())

print("\nПервые 5 строк CSV:")
print(df_csv.head())
````
